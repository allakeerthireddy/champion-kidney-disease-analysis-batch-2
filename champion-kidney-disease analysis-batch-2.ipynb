# Normalization of data

```python
from sklearn import preprocessing
# Normalize the numeric variables in the dataframe
minmax = preprocessing.MinMaxScaler(feature_range=(0,1))
minmax.fit(X).transform(X)
```

```python
from sklearn import model_selection,neighbors
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size  = 0.2)
```

```python
# Fitting Classifier to the Training set
clf = neighbors.KNeighborsClassifier()
clf.fit(X_train,y_train)
```

```python
# Predict on test data
y_pred = clf.predict(X_test)
```

```python
print("Actual kidney disease patients:")
print(y_test.values)
```

```python
print("\n Predicted kidney disease patients:")
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,recall_score,roc_auc_score,confusion_matrix
print("\n Accuracy score :%f"%(accuracy_score(y_test,y_pred)*100))
```

```python
print("Recall score:%f"%(recall_score(y_test,y_pred)*100))
```

```python
print("ROC score:%f\n"%(roc_auc_score(y_test,y_pred)*100))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
x = confusion_matrix(y_test,y_pred)
TP = x[0][0]
FP = x[0][1]
FN = x[1][0]
TN = x[1][1]
Specificity = TN/(TN+FP)
Sensitivity = TP/(TP+FN)
print('Sensitivity of KNN:',Sensitivity)
print('Specificity of KNN:',Specificity)
```
